# Plotly Assignment

```python
# Q1

import seaborn as sns
import plotly.express as px

titanic_data = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express for age and fare columns
scatter_plot = px.scatter(titanic_data, x="age", y="fare", title="Titanic: Age vs Fare Scatter Plot")
scatter_plot.show()
```

```python
# Q2

# Load the tips dataset from Plotly
tips_data = px.data.tips()

# Create a box plot using Plotly Express
box_plot = px.box(tips_data, x='day', y='total_bill', color='day', title='Total Bill Distribution by Day')
box_plot.show()
```

```python
# Q3

hist_plot = px.histogram(tips_data,x='sex',y='total_bill',title='Histogram between sex and total_bill',
                         pattern_shape='smoker',color='day')
hist_plot.show()
```

```python
# Q4
iris_data = px.data.iris()
sct_mat = px.scatter_matrix(iris_data,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                            color='species',title="Scatter Matrix Plot of Iris Dataset by Species")
sct_mat.show()
```

```python
# Q5

# In Plotly Express, the distplot function has been deprecated in favor of the histogram function for creating histograms.

histogram = px.histogram(tips_data, x='total_bill', title='Histogram of Values')
histogram.show()
```
